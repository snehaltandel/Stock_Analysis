```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split as tts
file_path = "/Users/snehalbhaitandel/Projects/stock_analysis/MRNS_60_minute_data_2018-12-23 17:42:53.769242/part-00000-7de0261a-fbc5-4a73-a096-fe4dd774fae3-c000.csv"
df = pd.read_csv(file_path)
```

```python
X = df[['Open', 'High', 'Low']]
y = df['Close']
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
print('Coef:', lm.coef_)
print('Intercept:', lm.intercept_)
```

```python
predictions = lm.predict(X_test)
sns.scatterplot(y_test, predictions)
```

```python
sns.distplot((y_test-predictions),bins=30)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
```

```python
print('MSE:', metrics.mean_squared_error(y_test, predictions))
```

```python
print('RMS:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('EVS:', metrics.explained_variance_score(y_test, predictions))
```
